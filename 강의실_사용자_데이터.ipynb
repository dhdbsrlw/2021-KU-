```python
import pandas as pd
import numpy as np

# 건물별 이동시간 최종.csv 파일 불러오기
building = pd.read_csv('/content/건물별 이동시간 최종.csv')
```

```python
from pandas import Series, DataFrame
import random 
from datetime import datetime, timezone

'''
콘센트 필요도: 숫자 1-5   ex) 1 - 필요하지 않음 ... 5 - 반드시필요 
책상 형태 (다인/개인) : 개인: 1, 다인: 0
의자 형태 (일체/분리) : 일체: 1, 분리: 0
랜드마크(건물) : 랜드마크 이름
시간 : isoformat
'''

# 사용자 번호 생성(이름 대용)
def generate_user(num):
  user = np.arange(num).tolist()
  return user

# 사용자 콘센트 필요도 생성
def generate_plug(user_num):
  outlet = random.choices([1, 2, 3, 4, 5], [1, 2, 3, 6, 5], k=user_num)   # 2번째 인자 == 가중치
  return outlet

# 사용자 선호 책상 형태 생성
def generate_desk(user_num):
  desk = random.choices([1, 0], [7, 3], k=user_num)
  return desk

# 사용자 선호 책상 형태 생성
def generate_chair(user_num):
  chair = random.choices([1, 0], [2, 8], k=user_num)
  return chair

# 사용자 위치 랜드마크(건물) 생성
def generate_building(user_num):
  location = []
  for i in range(user_num):
    building_idx = np.random.randint(len(building))
    location.append(building.iloc[building_idx]['building'])
  return location
    
# 사용자 시간 생성
def generate_time(user_num):  # 시간 형식: 2021-07-01T03:00:00+09:00
  total_time = []
  for i in range(user_num):
    year = 2019
    month = random.randint(1,12)
    if month in [1,3,5,7,8,10,12]:
      day = random.randint(1,31)
    elif month == 2:
      day = random.randint(1,28)
    else :
      day = random.randint(1,30)
    hour = random.randint(9,23)
    minute = random.randint(0,59)
    second = random.randint(0,59)         
    time = datetime(year, month, day, hour, minute, second, tzinfo=timezone.utc).isoformat()
    total_time.append(time)
  print(total_time)
  return total_time


# 생성할 사용자 수 =  5000명
user_num = 5000

# 가상 사용자 데이터 생성
user = generate_user(user_num)
location = generate_building(user_num)
time = generate_time(user_num)
outlet = generate_plug(user_num)
desk = generate_desk(user_num)
chair = generate_chair(user_num)


# 가상 사용자 데이터프레임 만들기
col = {
       'user_index': user, 
       'building': location,
       'time': time,
       'plugs': outlet,
       'table': desk, 
       'chair': chair
      }

classroom_user_data = pd.DataFrame(col)
classroom_user_data

```
